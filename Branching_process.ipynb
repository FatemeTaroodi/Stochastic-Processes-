```python
import random
import matplotlib.pyplot as plt

# مقادیر پارامترهای a0, a1, a2
a0 = 0.3  # احتمال اینکه فرد هیچ فرزندی نداشته باشد
a1 = 0.5  # احتمال اینکه فرد یک فرزند داشته باشد
a2 = 0.2  # احتمال اینکه فرد دو فرزند داشته باشد

# تابعی برای شبیه‌سازی تعداد فرزندهایی که یک فرد تولید می‌کند
def number_of_children():
    rand = random.random()  # یک عدد تصادفی بین 0 و 1
    if rand < a0:
        return 0
    elif rand < a0 + a1:
        return 1
    else:
        return 2

# تابع شبیه‌سازی درخت فرایند شاخه‌ای
def branching_process(initial_population):
    population = [initial_population]  # لیست ذخیره تعداد افراد در هر نسل
    current_population = [initial_population]  # شروع از یک فرد

    # ادامه شبیه‌سازی تا زمانی که همه افراد منقرض شوند
    while sum(current_population) > 0:
        next_population = []  # لیست برای ذخیره جمعیت نسل بعدی

        # برای هر فرد در نسل جاری، تعداد فرزندهای آن محاسبه می‌شود
        for individual in current_population:
            for _ in range(individual):
                next_population.append(number_of_children())  # تعداد فرزندهای فرد فعلی

        # تعداد افراد نسل بعدی
        population.append(sum(next_population))
        current_population = next_population  # نسل جدید

    return population

# تعداد افراد اولیه
initial_population = 1  # شروع از یک فرد

# تعداد دفعات شبیه‌سازی
num_simulations = 5  # تعداد اجراهای شبیه‌سازی

# ایجاد یک نمودار برای هر شبیه‌سازی
plt.figure(figsize=(10, 6))

for i in range(num_simulations):
    # شبیه‌سازی فرایند برای هر بار اجرا
    population_over_time = branching_process(initial_population)

    # رسم نمودار برای این شبیه‌سازی
    plt.plot(population_over_time, label=f'Simulation {i + 1}')

# اضافه کردن عنوان و برچسب‌ها به نمودار
plt.xlabel('Generation')
plt.ylabel('Population Size')
plt.title(f'Branching Process Simulations (Initial Population: {initial_population})')
plt.legend()
plt.show()

# نمایش نتایج
for i in range(num_simulations):
    print(f"Population sizes for Simulation {i + 1}: {branching_process(initial_population)}")
```

```python
import random

def simulate_extinction(a0, a1, a2, max_generations=1000):
    # شروع با یک موجود
    population = 1
    generations = 0

    # شبیه‌سازی تکامل موجودات
    while population > 0 and generations < max_generations:
        new_population = 0
        for _ in range(population):
            # تصمیم‌گیری هر موجود برای تولید فرزند
            r = random.random()
            if r < a0:
                # موجود از بین می‌رود
                continue
            elif r < a0 + a1:
                # موجود یک فرزند تولید می‌کند
                new_population += 1
            else:
                # موجود دو فرزند تولید می‌کند
                new_population += 2

        population = new_population
        generations += 1

        # اگر تمام موجودات منقرض شدند
        if population == 0:
            return True  # انقراض رخ داده است

    return False  # انقراض رخ نداده است

# تعداد شبیه‌سازی‌ها
num_simulations = 10000
extinction_count = 0

# شبیه‌سازی انقراض در تعداد زیادی تکرار
for _ in range(num_simulations):
    if simulate_extinction(0.3, 0.5, 0.2):
        extinction_count += 1

# محاسبه احتمال انقراض
extinction_probability = extinction_count / num_simulations
print(f"احتمال انقراض: {extinction_probability:.4f}")
```
